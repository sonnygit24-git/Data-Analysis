# 문제) 각 사용자 별 평균 평점 계산하기

```python
#user_id에 대해 unique값만 먼저 뽑기
user_ids = np.unique(data[:,0])
```

```python
user_ids[:10]
```

```python
user_ids.shape
```

```python
#빈 리스트 생성
mean_rating_by_user_list = []

for user_id in user_ids:
    data_for_user = data[data[:,0] == user_id, :]  #data_for_user는 특정 user_id 값들의 array를 저장한 변수 (unique값별로 저장)
    mean_rating_for_user = data_for_user[:, 2].mean()  # rating의 평균 
    mean_rating_by_user_list.append([user_id, mean_rating_for_user]) # 빈리스트에 값 넣기
```

```python
data_for_user[:10]
```

```python
mean_rating_by_user_list[:10]
```

```python
mean_rating_by_user_list_array = np.array(mean_rating_by_user_list, dtype=np.float64)
```

```python
mean_rating_by_user_list_array.shape
```

# 파일에 writing하기

```python
np.savetxt("./data/movielens/mean_rating_by_user.csv", mean_rating_by_user_list_array, fmt="%.3f", delimiter=',')
```
