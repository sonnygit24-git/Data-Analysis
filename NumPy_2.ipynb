```python
data = np.genfromtxt('./data/movielens/ratings.dat', delimiter = '::', dtype=np.int64)
data
```

```python
np.mean(data[:, 2])
```

```python
data[:, 2].mean()
```

```python
user_ids = np.unique(data[:,0])
user_ids
```

```python
user_ids.shape
```

```python
# 빈 리스트 생성
mean_rating_by_user_list = []

for user_id in user_ids:
    data_for_user = data[data[:,0] == user_id, :] # 특정 user_id 값들의 array를 저장한 변수(unique값별로 지정)
    mean_rating_for_user = data_for_user[:, 2].mean() # rating의 평균
    mean_rating_by_user_list.append([user_id, mean_rating_for_user]) # 빈 리스트에 값 넣기
```

```python
mean_rating_by_user_list[:10]
```
