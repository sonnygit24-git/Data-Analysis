##  데이터 타입
numpy에서 사용 가능한 데이터형(dtype) 정리  
  
**int8, int16, int32, int64** : 부호가 있는 [8, 16, 32, 64]비트 정수  
**unit8, unit16, unit32, unit64** : 부호가 없는 [8, 16, 32, 64]비트 정수  
**float16, float32, float64, float128** : [16, 32, 64, 128]비트 실수  
**complex64, complex128, complex256, complex512** : [64, 128, 256, 512]비트 복소수  

int 부호가 있는 정수  
unit 부호가 없는 정수  
float 실수  
complex 복소수  
bool 불리언 (True 또는 False)  
object python object 형  
string 문자열  
unicode 유니코드 문자열  

```python
float_arr = arr.astype(np.float64)
float_arr
```

```python
float_arr.dtype
```

```python
arr1 = np.array([[1,2,3], [4,5,6]], dtype=np.float64)
arr2 = np.array([[7,8,9], [10,11,12]], dtype=np.float64)
```

```python
arr1
```

```python
arr2
```

```python
arr1 + arr2
```

```python
arr1 * 2
```

```python
arr1 ** 2
```

```python
1 / arr1
```
